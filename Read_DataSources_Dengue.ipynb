```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
ds_raca = pd.read_csv('DataSource\Raca_Casos_2020.csv',  sep=';')
ds_raca
```

```python
ds_escolaridade = pd.read_csv('DataSource\Escolaridade_Casos_2020.csv',  sep=';')
ds_escolaridade
```

```python
ds_dengue = ds_escolaridade.merge(ds_raca, on='Reg.Metropolit/RIDE de notific')

ds_dengue
```

## Análise Exploratória

```python
#Dristribuiçãod dos casos por Raça

# make data
x = ds_raca.iloc[12][1:7]
y = ['Ign/Branco', 'Branca', 'Preta', 'Amarela', 'Parda', 'Indigena']

colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))

# plot
fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, labels=y)



plt.show()
```

```python
#Dristribuiçãod dos casos por escolaridade
x = ds_escolaridade.iloc[12][2:12]
y = ['Analfabeto', '1 a 4 serie incompleta do EF', '4 serie completa do EF', '5 a 8 serie incompleta do EF', 
     'Ensino fundamental completo', 'Ensino medio incompleto', 'Ensino medio completo', 'Educacao superior incompleta',
    'Educacao superior completa', 'Nao se aplica']

colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))

# plot
fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, labels=y)



plt.show()
```
