```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Casos de Dengue em Sorocaba: Predição e busca por anomalias 

#### Como objeto de estudo vamos observar a distribuição dos casos de dengue em na região de Sorocaba e tentar identificar anomalias e construir um modelo para prever os casos ao longo dos anos

```python
def GetOnlySorocabaData(dataSource):
    return dataSource.iloc[10][2:12]

def PlotPizzaChart(dataSource):
    x = dataSource
    y = ['Analfabeto', '1 a 4 serie incompleta do EF', '4 serie completa do EF', '5 a 8 serie incompleta do EF', 
     'Ensino fundamental completo', 'Ensino medio incompleto', 'Ensino medio completo', 'Educacao superior incompleta',
    'Educacao superior completa', 'Nao se aplica']

    colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))

    # plot
    fig, ax = plt.subplots()
    ax.pie(x, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, labels=y)



    plt.show()

def GetTotalSorocaba(dataSource):
    return dataSource.iloc[10][12]

def GetTotalSuperiorCompleto(dataSource):
    return dataSource.iloc[10][10]
```

## Ano 2016

```python
ds_escolaridade_2016 = pd.read_csv('DataSource\Escolaridade_2016.csv',  sep=';')
ds_escolaridade_2016
```

### Sorocaba

```python
ds_sorocaba_2016 = GetOnlySorocabaData(ds_escolaridade_2016)
ds_sorocaba_2016
```

```python
PlotPizzaChart(ds_sorocaba_2016)
```
